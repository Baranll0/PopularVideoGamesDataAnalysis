## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import ast
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Load Data

```python
df = pd.read_csv("backloggd_games.csv")
```

##  Step 3:Basic Information

```python
df.columns
df.info()
df.describe()
```

##  Step 4:Data Preprocessing

```python
#Convert 'Release_Date' from string to datetime and extract year
df['Release_Date'] = df['Release_Date'].replace('TBD', np.nan)
df['Release_Date'] = pd.to_datetime(df['Release_Date'], format="%b %d, %Y")
df['Year'] = df['Release_Date'].dt.year
```

## Step 5: Exploratory Data Analysis

##### Number of games released per year

```python
#Number of Games released per year
game_per_year=df.groupby('Year').size()
plt.figure(figsize=(10,6))
game_per_year.plot(kind='line')
plt.title("Number of Games Released per Year")
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.show()
```

##### Top Developers by Total Plays

```python
#Top Developers by Total Plays
def plays_to_numeric(play_str):
    #Convert 'Plays' string values to numeric
    if isinstance(play_str, str):
        if 'K' in play_str:
            return float(play_str.replace('K', '')) * 1000
        elif 'M' in play_str:
            return float(play_str.replace('M', '')) * 1000000
        else:
            return float(play_str)
    else:
        return play_str
df['Plays'] = df['Plays'].apply(plays_to_numeric)
df['Developer'] = df['Developers'].str.split(', ').str.get(0)
df['Developer'] = df['Developer'].str.replace('[', '', regex=False)
df['Developer'] = df['Developer'].str.replace(']', '', regex=False)
df = df[df['Developer'] != '']

# Her gelitirici iin oyunlarn toplam oyun saysn hesaplama
popularity = df.groupby('Developer')['Plays'].sum()

# En popler gelitiricileri sralama (yksekten de)
popularity = popularity.sort_values(ascending=False)

top_10_popular = popularity.head(20)

plt.figure(figsize=(15, 8))
sns.barplot(x=top_10_popular.index, y=top_10_popular.values, color='#00A0FF')
plt.title('Top 10 Popular Developers by Total Plays')
plt.xlabel('Developer')
plt.ylabel('Total Plays')
plt.xticks(rotation=50)
plt.show()
```

##### Most Popular Genres

```python
df['Genres']=df['Genres'].apply(ast.literal_eval)
df=df[df['Genres'].apply(lambda x:x!=[])]
genres = df.explode('Genres')
genre_counts = genres['Genres'].value_counts()
top_20_genres=genre_counts[:20]
plt.figure(figsize=(15,10))
sns.barplot(x=top_20_genres.index,y=top_20_genres.values,color='#00A0FF')
plt.title('Game Genre Popularity')
plt.xlabel('Genre')
plt.ylabel('Total Games')
plt.xticks(rotation=90)
plt.show()
```

##### Average Game Ratings by Platform

```python
df['Platforms'] = df['Platforms'].apply(lambda x: eval(x))
df_exploded = df.explode('Platforms')
df_exploded['Rating'] = df_exploded['Rating'].fillna(df_exploded['Rating'].mean())
df_exploded = df_exploded.dropna()

platform_ratings = df_exploded.groupby('Platforms')['Rating'].mean()
top_platforms = platform_ratings.sort_values(ascending=False).head(50)

plt.figure(figsize=(10, 8))
sns.barplot(x=top_platforms.values, y=top_platforms.index, palette='viridis')
plt.title('Average Game Ratings by Platform')
plt.xlabel('Average Rating')
plt.ylabel('Platform')
plt.show()
```

##### Correlation between Number of Plays and Ratings

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
correlation=df['Plays'].corr(df['Rating'])

plt.figure(figsize=(10,8))
plt.scatter(df['Plays'],df['Rating'],alpha=0.5)
plt.title('Correlation betweetn Number of Plays and Ratings')
plt.xlabel('Number of Plays')
plt.ylabel('Rating')
plt.show()
```

##### Top Developers by Average Game Rating

```python
df['Developers']=df['Developers'].apply(lambda x: eval(x) if isinstance(x,str) else x)
df_explode=df.explode('Developers')
developer_ratings= df_exploded.groupby('Developers')['Rating'].mean()
top_developer=developer_ratings.idxmax()


df['Developers']=df['Developers'].apply(lambda x: top_developer if top_developer in x else x[0])
top_developers=developer_ratings.sort_values(ascending=True).head(50)


plt.figure(figsize=(15,15))
sns.barplot(y=top_developers.index,x=top_developers.values,palette='viridis')
plt.title('Top 50 Developers by Average Game Rating')
plt.xlabel('Average Rating')
plt.ylabel('Developer')
plt.show()
```

##### Average Wishlist Additions per Rating

```python
def convert_k_to_thousands(s):
    s=s.replace('K','')
    if s:
        return float(s)*1000
    else:
        return 0.0
    
df['Wishlist']=df['Wishlist'].apply(convert_k_to_thousands)
average_wishlist_additions=df.groupby('Rating')['Wishlist'].mean()

plt.figure(figsize=(10,6))
plt.scatter(average_wishlist_additions.index,average_wishlist_additions.values,color='blue')
plt.title('Average Wishlist Additions per Rating')
plt.xlabel('Rating')
plt.ylabel('Average WiShlist Additions')
plt.show()
```

##### Average Ratings bt Game Genre

```python
genres = df.explode('Genres')
average_rating_by_genre=genres.groupby('Genres')['Rating'].mean()
average_rating_by_genre_sorted = average_rating_by_genre.sort_values(ascending=False)

colors = plt.cm.viridis(range(len(average_rating_by_genre_sorted)))


plt.figure(figsize=(12, 6))
plt.bar(average_rating_by_genre_sorted.index, average_rating_by_genre_sorted, color=colors)
plt.xlabel('Genres')
plt.ylabel('Average Rating')
plt.title('Average Ratings by Game Genre')
plt.xticks(rotation=90)
plt.show()
```

##### Average Number of Players by Genre

```python
genres_df = df.explode('Genres')
average_plays_by_genre = genres_df.groupby('Genres')['Plays'].mean()

genre_plays = average_plays_by_genre.sort_values(ascending=False)

plt.figure(figsize=(10, 8))
plt.barh(genre_plays.index, genre_plays.values)
plt.xlabel('Average Number of Players')
plt.ylabel('Genres')
plt.title('Average Number of Players by Genre')
plt.show()
```

##### Top Games by Number of Reviews

```python
review_counts = df.groupby('Title')['Reviews'].count().sort_values(ascending=False)
top_20_reviews = review_counts.head(50)

plt.figure(figsize=(15, 15))
sns.barplot(x=top_20_reviews.values, y=top_20_reviews.index, palette='viridis')
plt.xlabel('Number of Reviews')
plt.ylabel('Title')
plt.title('Top 50 Games by Number of Reviews')
plt.show()
```
